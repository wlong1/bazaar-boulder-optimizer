```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import json
import glob
```

```python
all_files = glob.glob('*.json')

runs = []
for file in all_files:
    with open(file, 'r') as f:
        runs.append(json.load(f))
```

```python
all_data = [item for run in runs for item in run]
min_val = min(all_data)
max_val = max(all_data)

bins = range(82, 98)

for run in runs:
    plt.hist(run, bins=bins, range=(min_val, max_val), alpha=0.5, 
             density=True)
    plt.xlabel('Ticks')
    plt.ylabel('Density')
    plt.title(f'DW, SC, RB, B, CW - {len(run)} Runs')
    plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    plt.show()
```
